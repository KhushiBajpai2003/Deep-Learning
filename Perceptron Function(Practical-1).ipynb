### **Perceptron Function**

```python
x_input=[0.1,0.5,0.6]
w_weight=[0.4,0.3,0.6]
threshold=0.5
```

```python
def step(weighted_sum):
  if weighted_sum>threshold:
    return 1
  else:
    return 0
```

```python
def preceptron():
  weighted_sum=0
  for x,w in zip(x_input,w_weight):
    weighted_sum+=x*w
    print(weighted_sum)
  return step(weighted_sum)
```

```python
output=preceptron()
print("Output:",output)
```

# **AND,OR,NOT GATES**

```python
def step_function(x):
    # Step function (threshold function)
    return 1 if x >= 0 else 0

def perceptron_AND(x1, x2):
    # AND gate perceptron
    w1, w2, threshold = 0.5, 0.5, 0.7
    weighted_sum = w1 * x1 + w2 * x2
    return step_function(weighted_sum)

def perceptron_OR(x1, x2):
    # OR gate perceptron
    w1, w2, threshold = 0.5, 0.5, 0.3
    weighted_sum = w1 * x1 + w2 * x2
    return step_function(weighted_sum)

def perceptron_NOT(x):
    # NOT gate perceptron
    w, threshold = -1, -0.5
    weighted_sum = w * x
    return step_function(weighted_sum)

# Testing the perceptrons
input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("AND Gate:")
for x1, x2 in input_data:
    output = perceptron_AND(x1, x2)
    print(f"{x1} AND {x2} = {output}")

print("\nOR Gate:")
for x1, x2 in input_data:
    output = perceptron_OR(x1, x2)
    print(f"{x1} OR {x2} = {output}")

print("\nNOT Gate:")
for x in [0, 1]:
    output = perceptron_NOT(x)
    print(f"NOT {x} = {output}")
```

```python
def step_function(x):
    return 1 if x >= 0 else 0

def perceptron_AND(x1, x2):
    w1, w2, bias = 0.5, 0.5, -0.7
    weighted_sum = w1 * x1 + w2 * x2 + bias
    return step_function(weighted_sum)

def perceptron_OR(x1, x2):
    w1, w2, bias = 0.5, 0.5, -0.3
    weighted_sum = w1 * x1 + w2 * x2 + bias
    return step_function(weighted_sum)

def perceptron_NOT(x):
    w, bias = -1, 0.5
    weighted_sum = w * x + bias
    return step_function(weighted_sum)

input_data = [(0, 0), (0, 1), (1, 0), (1, 1)]

print("AND Gate:")
for x1, x2 in input_data:
    output = perceptron_AND(x1, x2)
    print(f"{x1} AND {x2} = {output}")

print("\nOR Gate:")
for x1, x2 in input_data:
    output = perceptron_OR(x1, x2)
    print(f"{x1} OR {x2} = {output}")

print("\nNOT Gate:")
for x in [0, 1]:
    output = perceptron_NOT(x)
    print(f"NOT {x} = {output}")
```
